```python
import pandas as pd
import pandas.io.json as pd_json
import os

pd.set_option('display.max_colwidth', -1) # display entire text
```

```python
FILE_NAME = "IDM_NM.csv"
OUTS_DIR = "datos.gob.mx/utiles"
file_path = "{}/{}".format(OUTS_DIR, FILE_NAME)
abs_file_path = os.path.abspath(file_path)
```

```python
df = pd.read_csv(abs_file_path, encoding="utf-8")
```
